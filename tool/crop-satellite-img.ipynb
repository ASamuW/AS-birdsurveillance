# Grid the cropped picture

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def draw_grid_with_labels(img, step_size=500):
    # Image dimensions
    height, width, _ = img.shape

    # Create coordinate ranges
    # x_ticks = np.arange(0, width , 500 if step_size < 500 else step_size)
    # y_ticks = np.arange(0, height, 500 if step_size < 500 else step_size)
    x_ticks = np.arange(0, width , step_size)
    y_ticks = np.arange(0, height, step_size)

    # Plotting the image
    fig, ax = plt.subplots(figsize=(20,12))
    ax.imshow(img)

    # Set ticks
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)

    # Set tick labels
    ax.set_xticklabels([str(x) for x in x_ticks])
    ax.set_yticklabels([str(y) for y in y_ticks])

    # Draw grid lines
    ax.grid(True, color='blue', linestyle='-', linewidth=1, which='both')
    
    fig.savefig("test.jpg", format='jpg')
    
    # Show the plot
    plt.show()

# Specify the path to your image
image_path = 'pictures/gep/izumi/201701.jpg'
# Load the image
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Define grid properties
step_size = 200  # Distance between grid lines
y_max = img.shape[0]
x_max = img.shape[1]

# Set range
x1, y1 = 0, 0
x2, y2 = x_max,y_max

# Crop the image
img = img[y1:y2, x1:x2]
draw_grid_with_labels(img, step_size)
```

```python
# Draw all cropped images
date_of_picture = image_path.split("/")[3].split(".")[0]
max_range_i = int(x_max/step_size)
max_range_j = int(y_max/step_size)
idx=0
for j in range(max_range_j):
    for i in range(max_range_i):
        x1 = i*step_size
        y1 = j*step_size
        x2 = (i+1)*step_size
        y2 = (j+1)*step_size
        # Crop a image
        img_crop = img[y1:y2,x1:x2]
        img_tmp = cv2.cvtColor(img_crop, cv2.COLOR_BGR2RGB)  # Convert color from BGR to RGB
        idx=idx+1
        # Save the image
        cv2.imwrite('pictures/gep_cropped/image_' + date_of_picture + "_" +str(step_size)+'_' + str(i) + '_' + str(j) +'.jpg',img_tmp)
```
